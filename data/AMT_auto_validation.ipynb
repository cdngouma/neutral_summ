### 3-Check for originality
Discard summaries that are copies of the input reviews

```python
def get_ngrams(text, n=5):
    grams = ngrams(text.split(), n)
    res = []
    for gram in grams:
        res.append(" ".join(gram))
    return set(res)
```

```python
rev_cols = [f"review_{i+1}" for i in range(0, 15)]
prod_ngrams = dict()
for prod_id in valid_prod_ids:
    rev_ngrams = []
    for rev_id in rev_cols:
        text = reviews.loc[reviews["prod_id"] == prod_id, rev_id].values[0]
        res = get_ngrams(text=text, n=6)
        rev_ngrams.append(res)
    prod_ngrams[prod_id] = rev_ngrams
```

```python
def check_for_copies(df, rev_ngrams, n=6):
    keep_idx = []
    discard_idx = []
    report = dict()
    for row in df.iterrows():
        idx = row[0]
        summary = row[1]["Answer.summary"]
        summ_ngrams = get_ngrams(summary, n)
        overall = 0
        for grams in  rev_ngrams:
            overlap = len(grams.intersection(summ_ngrams))
            if overlap >= 2:
                id_ = f"{row[1]['WorkerId']}.{row[1]['AssignmentId']}"
                if report.get(id_) is None:
                    report[id_] = {
                        "summ": summary,
                        "overlaps": []
                    }
                report[id_]["overlaps"].append(grams)
                overall = 15
                break
            else:
                overall += overlap
        
        if overall >= 2:
            discard_idx.append(idx)
        else:
            keep_idx.append(idx)
    
    return keep_idx, discard_idx, report
```

```python
def discard_copies(df, prod_ids, prod_ngrams, n=6):
    keep_idx = []
    discard_idx = []
    report = dict()
    for prod_id in prod_ids:
        summ_df = df.loc[df["Input.prod_id"] == prod_id].copy()
        k,d,r = check_for_copies(summ_df, prod_ngrams.get(prod_id), n=n)
        keep_idx.extend(k)
        discard_idx.extend(d)
        report.update(r)
    keep = df.loc[keep_idx]
    discard = df.loc[discard_idx]
    return keep, discard, report
```

```python
keep, discard, report = discard_copies(results_batch, valid_prod_ids, prod_ngrams, n=6)
```
