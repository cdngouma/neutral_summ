The purpose of this notebook is to transform our test data in the proper format for Amazon Mechanical Turk.

```python
import pandas as pd
```

```python
test = pd.read_csv("./data/raw/20220420_amazon_reviews_test.csv")
test.shape
```

```python
test.head()
```

```python
test["prod_id"].unique().shape
```

```python
df = test[["prod_id", "review"]]
df.head()
```

```python
df_grouped = df.groupby(["prod_id"], as_index=False).agg(lambda x: "<<SEP>>".join(x.values))
df_grouped.shape
```

```python
df_grouped.head()
```

```python
df_grouped = df_grouped.join(df_grouped["review"].str.split("<<SEP>>", expand=True).rename(columns={i: f"review_{i+1}" for i in range(0,15)}))
```

```python
df_grouped = df_grouped.drop(columns=["review"])
df_grouped.head()
```

```python
df_grouped.shape
```

```python
df_grouped.to_csv("./AMT_neutral_summ_test_data.csv", index=False)
```
